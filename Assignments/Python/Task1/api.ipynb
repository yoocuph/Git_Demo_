```python
# write a function that will double the following numbers and store them in another variable double
df = pd.DataFrame({'number': [1,2,3,4,5]})

def double_double(x):
    d = x * 2
    return d

df['double'] = df['number'].apply(double_double)
#df
print(df)
#display(df)
```

```python
# Make a function that returns "Even" or "Odd", then apply it

df = pd.DataFrame({'value': [10, 15, 22, 7, 3]})

def even_odd(x):
    if x % 2 == 0:
        return "even"
    else:
        return "odd"
    
df['doub'] = df['value'].apply(double_double)
df['checkeven'] = df['value'].apply(even_odd)
df
```

```python
# Work with text â€” make everything uppercase after you must have stripped all the unnessesary spaces

df_g = pd.DataFrame({'name': ['alice ', ' bob ', 'cha rlie ', ' dave ']})
df_g

def make_upper(x):
    a = x.strip().upper().replace(" ", "")
    return a

df_g['namee'] = df_g['name'].apply(make_upper)
df_g
```

```python
# Grade Calculator

df = pd.DataFrame({'score': [95, 82, 75, 60, 45, 88]})

def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else: 
        return "F"
    
df['grade'] = df['score'].apply(get_grade)
df
```

```python
# Categorize Age Groups
df = pd.DataFrame({'age': [12, 17, 25, 45, 67, 8, 33]})

def age_group(age):
    if age < 13:
        return "Child"
    elif 13 <= age < 18:
        return "Teen"
    elif 18 <= age < 65:
        return "Adult" 
    elif age >= 65:
        return "Senior"
    else: 
        return None
df['category'] =  df['age'].apply(age_group)
df
```

```python
# Extract First Letter of Each Word

df = pd.DataFrame({'full_name': ['John Doe', 'Alice Smith', 'Bob Johnson', 'Eve Brown']})

def get_initials(name):
    parts = name.split()
    return ".".join(p[0].upper() for p in parts)

df['initials'] = df['full_name'].apply(get_initials)
df

```
